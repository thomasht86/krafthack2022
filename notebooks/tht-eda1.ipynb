```python
df1 = pd.read_parquet("../container-data/input_dataset-1.parquet")
df2 = pd.read_parquet("../container-data/input_dataset-2.parquet")
```

```python
pred_cols = ["Bolt_"+str(x)+"_Tensile" for x in range(1,7)]+list(pred_df.columns)
pred_cols
```
